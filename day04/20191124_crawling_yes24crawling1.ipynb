```python
# 라이브러리 임포트
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import codecs
```

```python
# 크롬드라이버 세팅(외부폴더에 있는것 가져오기)
driver = webdriver.Chrome("C:/Users/SJ/data1112/chromedriver.exe")

# yes24 서점 베스트셀러 페이지로 이동
driver.get("http://www.yes24.com/24/Category/BestSeller")

# 딜레이(2초)
time.sleep(2)

# 전체 소스코드 가져오기
source = driver.page_source
print(source)
```

```python
# 소스코드 파싱하기
s1 = BeautifulSoup(source, "html.parser")

# 태그명이 li인 요소만 가져오기
s_book = s1.find_all("ol")

# 몇 개인지 측정
len(s_book)
```

```python
# s_book에서 li태그인 요소만 다시 조회
len(s_book[0].find_all("li", class_=class_list))
```

```python
# 문제
class_list = []
# 위 리스트에 .append()를 이용하고
# for문을 이용해서 첫 번째 바퀴에는 "num%s" % 1.... 과 같이
# num40까지 반복문을 이용해 append 시켜서 class_list 변수에 num1 ~ num40을 채워보세요.
# 단 num19와 num20은 홈페이지를 참조해 홈페이지에 나온 패턴으로 만들어주세요.
for i in range(1, 41):
    if (i == 19) or (i == 20):
        class_list.append("num%s_line" % i)
    else:
        class_list.append("num%s" % i)
print(class_list)
```

```python
# 40개의 li 내부에서 제목, 저자, 가격순으로 자료를 뽑아서 나열해주세요.
s_book_li = s_book[0].find_all("li", class_=class_list)

rank = 0
for item in s_book_li:
    rank+=1
    p = item.find_all("p")
    print("%s위 책 정보" % rank)
    print(p[2].text)
    print(p[3].text)
    print(p[4].text)
    print("="*50)
```

```python
# result3.txt 파일에 결과를 옮겨서 저장해주세요.
f = codecs.open("C:/Users/SJ/data1112/day04/result3.txt",
               encoding="utf-8", mode='w')

rank = 0
for item in s_book_li:
    rank+=1
    p = item.find_all("p")
    f.write("%s위 책 정보\r\n" % rank)
    f.write(p[2].text+"\r\n")
    f.write(p[3].text+"\r\n")
    f.write(p[4].text+"\r\n")
    f.write("=\r\n"*50)
f.close()
```
