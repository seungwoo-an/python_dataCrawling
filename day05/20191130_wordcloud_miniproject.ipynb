## 본인이 크롤링해온 자료를 워드클라우드화 해 주세요

- 한글도 괜찮고 영문도 괜찮습니다.

- 워드클라우드화 시키는데 필요한 파일은 전부 같은 폴더에 둡니다.

-  배경 없이도 진행해보시고 단톡방에 올라온 한반도 사진을 이용해서도 워드클라우드화 해 보세요.

```python
from wordcloud import WordCloud, STOPWORDS

import matplotlib.pylab as plt

from konlpy.tag import Twitter
import nltk

import numpy as np
from PIL import Image
```

```python
#자료 오픈
articles=open('daumNews.txt',encoding='utf-8').read()
article=articles.replace('및','').replace('입력','').replace('내용','').replace('댓글','').replace('공감','').replace('관련','').replace('뉴스','').replace('등','').replace('것','').replace('보기','').replace('더','').replace('재','')
#배경자료
background = np.array(Image.open('kor_map.png'))

t=Twitter()
token = t.nouns(article)
#한국어 워드클라우딩은 단어 카운팅이 선행
count_voca = nltk.Text(token, name='단어카운팅')
voca = count_voca.vocab().most_common(150)#상위150
#한국어 폰트가 원드클라우드에 내장되어 있지 않으므로 글꼴을 지정해줌
font_path = 'c:/Windows/Fonts/malgun.ttf'
wordcloud = WordCloud(font_path=font_path, relative_scaling=0.2, background_color='white', mask=background
                     ).generate_from_frequencies(dict(voca))
#크롤링 자료  워드클라우드 출력
plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
