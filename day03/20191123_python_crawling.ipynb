#### BeautifulSoup를 활용한 텍스트 자료 가공

- selenium을 이용해 원하는 사이트의 전체 코드를 긁어왔다면 BeautifulSoup를 이용해서 원하는 자료만 남겨야 한다.
    - 즉, BeautifulSoup는 자료 수집과는 전혀 관련이 없는 라이브러리이다.
- 기본적으로 "파싱" 이라는 작업을 해야 뷰티플수프에게 내가 원하는 부분만 남겨달라고 요청할 수 있다.

```python
#BeautifulSoup(소스코드(이해해야 되는 내용),"html.parser")입력 시 파싱작업 완료
s1 = BeautifulSoup(source,"html.parser")
print(s1)
type(s1)
```

<b>Beautifulsoup 기본 명령어</b><br>
<hr color=black>
파싱한자료.find_all(“태그명”, id=“아이디명”, class_=“클래스명”)<br>
파싱한 자료 내부에서 원하는 태그명 및 아이디, 클래스에 해당하는
자료만 선별하는 코드.
이 코드를 활용해 원하는 텍스트자료가 있는 부분을 선택적으로 수집
할 수 있게 된다.<br>
만약 두 개 이상의 요소를 동일하게 수집하고 싶다면 리스트 처리해
find_all([“태그명1”, “태그명2”…]) 과 같이 처리한다.<br>
- 최대한 특정지을 수 있는 것을 집어야 함

```python
#타이틀 관련해서 가져오기
s_title = s1.find_all('div',class_='title')
#꺾쇠 여닫는 부분 다 삭제
titles=[]
for title in s_title[4:]:
    titles.append(title.text.strip())
s_author = s1.find_all('div',class_='author')
s_price = s1.find_all('div',class_='price')
authors=[]
prices=[]
for author in s_author:
    authors.append(author.text.strip()[0:3].replace('\n','').replace(' ',''))
for price in s_price:
    price = price.text.strip()
    price=price[price.find('원')-6:price.find('원')]
    prices.append(price.replace('\t',''))
file=[]
for i in range(len(titles)):
    print("%d위 작가 %s의 책 제목 : %s\n> 가격 : %s원\n"%(i+1,authors[i],titles[i],prices[i]))
```

### codecs 라이브러리

- 파이썬 3.5버전 들어서 내장 라이브러리로 바뀜(예전에는 pip로 설치해야 했었음)
- 파이썬으로 텍스트파일을 제어할 수 있도록(읽어오기, 쓰기) 도와줌
- 콘솔창에 출력된 내용을 txt파일로 옮겨서 출력할때 사용
<br><br>
- 특이사항
    - 개행은 \r\n으로 처리함
    - mode => w(기존에 있던 자료 없애고 새 파일 입력)
    - mode => a(기존에 있던 자료에 이어서 계속 입력)
    - mode => r(텍스트파일에 있던 내용 읽어오기)

```python
#라이브러리 import
import codecs
```

```python
#f 변수가 text파일 그 자체처럼 사용됨
#.open(파일경로(파일 없으면 생성, 있으면 덮어씌우기, encoding='인코딩', mode='모드'))
f = codecs.open('C:/Users/SJ/data1112/day03/test.txt',encoding='utf-8', mode='w')
# .write(적을 내용)을 실행하면 텍스트파일 내부에 문자열이 입력된다.
for i in range(10):
    f.write('hello world\r\n')
#파일 닫기(항상 마지막줄에 배치)
f.close()
```
